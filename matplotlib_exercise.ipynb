```python

import numpy as np  
import matplotlib.pyplot as plt  
  
X = ['saiee','vaishu','samiksha','shruti'] 
Age = [20,30,21,45] 
Height = [120,110,125,135] 
  
X_axis = np.arange(len(X)) 
  
plt.bar(X_axis -0.2, Age,0.4, label = 'Age') 
plt.bar(X_axis +0.2,Height,0.4, label = 'Height') 
  
plt.xticks(X_axis, X) 
plt.xlabel("Names") 
plt.ylabel("") 
plt.title("Age and Height of Girls") 
plt.legend() 
plt.show() 
```

```python
#Barh plot

y = ['saiee','vaishu','samiksha','shruti'] 
Age = [20,30,21,45] 
Height = [120,110,125,135] 
  
y_axis = np.arange(len(y)) 
  
plt.barh(y_axis -0.2, Age,0.4, label = 'Age') 
plt.barh(y_axis +0.2,Height,0.4, label = 'Height') 
  
plt.yticks(y_axis, y) 
plt.ylabel("Names") 
plt.xlabel("") 
plt.title("Age and Height of Girls") 
plt.legend() 
plt.show() 
```

```python
#scatter plot
x = ['saiee','vaishu','samiksha','shruti'] 
Age = [20,30,21,45] 
Height = [120,110,125,135] 
  
x_axis = np.arange(len(x)) 
  
plt.scatter( x_axis,Age, label = 'Age') 
plt.scatter(x_axis,Height, label = 'Height') 
  
plt.xticks(x_axis, x) 
plt.xlabel("Names") 
plt.ylabel("") 
plt.title("Age and Height of Girls") 
plt.legend() 
plt.show() 
```
